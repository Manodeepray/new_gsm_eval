```python
import openai
import json
import re

def evaluate_response(user_response, correct_answer, api_key):
    """
    Compare the user's response with the correct answer using the OpenAI API.

    :param user_response: The response given by the user (string).
    :param correct_answer: The correct answer from the JSON (string).
    :param api_key: OpenAI API key for authentication (string).
    :return: Evaluation result from the OpenAI API (string).
    """
    openai.api_key = api_key

    # Construct the prompt for the API
    prompt = (
        "Evaluate the user's response for correctness, completeness, and clarity "
        "compared to the correct answer provided. Provide a score out of 10 and detailed feedback.\n\n"
        f"Correct Answer: {correct_answer}\n"
        f"User Response: {user_response}\n\n"
        "Evaluation:"
    )

    try:
        # Make a call to the OpenAI API
        response = openai.ChatCompletion.create(
            engine="text-davinci-003",  # Specify the model you want to use
            prompt=prompt,
            max_tokens=200,
            temperature=0.5
        )

        # Extract and return the evaluation result
        evaluation = response.choices[0].text.strip()
        return evaluation

    except Exception as e:
        return f"An error occurred: {str(e)}"

def extract_number_after_hashes(input_string):
    """
    Extract the number after '###' in the input string.

    :param input_string: The input string to search (string).
    :return: The number after '###' or None if not found (int or None).
    """
    match = re.search(r"###\s*(\d+)", input_string)
    if match:
        return int(match.group(1))
    return None


api_key = "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"

# Evaluate the response
evaluation = evaluate_response(user_response, correct_answer, api_key)
print(evaluation)
```

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key=api_key,  # This is the default and can be omitted
)

def extract_number_after_hashes(input_string):
    """
    Extract the number after '###' in the input string.

    :param input_string: The input string to search (string).
    :return: The number after '###' or None if not found (int or None).
    """
    match = re.search(r"###\s*(\d+)", input_string)
    if match:
        return int(match.group(1))
    return None


def answer_checker(data):
    correct_answer = data['answer']
    question = data['question']
    user_response = str(data['response'][0])

    correct_answer = extract_number_after_hashes(correct_answer)

    message = [
            {
                "role": "system",
                "content": f'''
                 evaluate if the user_response is same as correct_answer \n
                 answer in correct or incorrect\n
                            question : {question}
                            correct_answer :{correct_answer} ,
                            user_response :{user_response}
                            Evaluation :
                            '''
            }
    ]

    chat_completion = client.chat.completions.create(
        messages=message,
        model="gpt-4o-mini",
    )
    print(chat_completion.choices[0].message.content)
    return chat_completion.choices[0].message.content


```

```python
import json
with open("llama3_1_8b_replaced_results_complete.json" , 'r') as f:
    data = json.load(f)

llama3_1_8b_replaced_results = [] 
for row in data:
    llama3_1_8b_replaced_results.append(answer_checker(row))
```

```python
count = 0
wrong = 0
for result in llama3_1_8b_replaced_results:
    count+=1
    if "incorrect" in str(result):
        wrong += 1
        
        
acc = (1-(wrong/count))
print("llama3_1_8b_replaced_results accuracy: ",acc)
        
```
